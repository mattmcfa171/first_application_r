```r
dataset = read.csv("regrex1.csv")
```

```r
print(dataset)
```

```r
model = lm(formula = x ~ y, data = dataset)
```

```r
print(model)
```

```r
library(ggplot2)
ggplot() +
    geom_point(aes(x = dataset$x, y = dataset$y),
            colour = 'red') +
    geom_line(aes(x = dataset$x, y = predict(model)),
            colour = 'blue') +
    ggtitle('X vs Y') +
    xlab('X') +
    ylab('Y')
```

```r
plot(x ~ y, data = dataset)
abline(model)
```

```r
library(ggplot2)
ggplot(dataset, aes(x = x, y = y)) +
    geom_point(color = 'red') +
    stat_smooth(method = "lm", se=FALSE, color = "blue")
```
